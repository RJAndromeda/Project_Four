```python
# Initial imports
import pandas as pd
from pathlib import Path
from sklearn import tree
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np


import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
from sklearn.linear_model import Lasso
import statsmodels.api as sm
```

```python
df1 = pd.read_csv('clean_Ballarat_slim.csv')
df2 = pd.read_csv("clean_Wangaratta_slim.csv")
df3 = pd.read_csv("clean_Ferny_Creek_slim.csv")
df4 = pd.read_csv("additional_Ferny_Creek_slim.csv")
cfa = pd.read_csv('cfa_grouped.csv')
```

```python
merged_df = df1.combine_first(df2)
merged_df = merged_df.combine_first(df3)
merged_df = merged_df.combine_first(df4)
# This adds in values from the other dataset to make up for missing data. Possibly not a great idea.
```

```python
merged_df.head()
```

```python
new_order = ['Date','Ball_Direction of maximum wind gust','Ball_Maximum temperature (°C)',\
             'Ball_Minimum temperature (°C)','Ball_Speed of maximum wind gust (km/h)',\
             'Fern_Direction of maximum wind gust','Fern_Maximum temperature (°C)',\
             'Fern_Minimum temperature (°C)','Fern_Speed of maximum wind gust (km/h)',\
             'Wang_Direction of maximum wind gust','Wang_Maximum temperature (°C)',\
             'Wang_Minimum temperature (°C)','Wang_Speed of maximum wind gust (km/h)']
merged_df = merged_df[new_order]
```

```python
merged_df
```

```python
merged_df.to_csv('merged_df.csv', sep=',', index=False)
```

```python
cfa
```

```python
CFA_dataset = pd.merge(cfa, merged_df, left_on="incident_datetime", right_on="Date", how='inner')
# going to only combine where there is a date in both columns
```

```python
CFA_dataset.tail()
```

```python
CFA_dataset = CFA_dataset.drop(columns=["District_no","incident_type_code","incident_type", "Date"])
```

```python
CFA_dataset = CFA_dataset.dropna()
```

```python
CFA_dataset.to_csv('CFA_dataset_clean.csv', sep=',', index=False)
```
