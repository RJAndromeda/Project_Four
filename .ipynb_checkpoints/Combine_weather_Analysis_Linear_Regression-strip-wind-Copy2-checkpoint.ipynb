```python
X = CFA_dataset.copy()
columns_dropped = ["Incident_count"]
X.drop(columns_dropped, axis=1, inplace=True)
X['CrossProduct'] = X['Ball_Maximum temperature (°C)']*X['Ball_Speed of maximum wind gust (km/h)']
X.head()
```

```python
ya = CFA_dataset["Incident_count"].ravel()
ya[:5]
```

```python
# Splitting into Train and Test sets
X_train, X_test, ya_train, ya_test = train_test_split(X, ya, random_state=78)
```

```python
# Creating StandardScaler instance
scaler = StandardScaler()
```

```python
# Fitting Standard Scaler
X_scaler = scaler.fit(X_train)
```

```python
# Scaling data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
print(X_train_scaled)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear = LinearRegression()
```

```python
linear.fit(X_train_scaled,ya_train)
```

```python
LinearRegression(copy_X=True, fit_intercept=True, n_jobs=1)
```

```python
print('Coefficients: \n', linear.coef_)
```

```python
predictions = linear.predict(X_test_scaled)
```

```python
# plt.scatter(ya_test,predictions)
# plt.xlabel('Y Test')
# plt.ylabel('Predicted Y')
result_df = pd.DataFrame({"Actual":ya_test, 'Predicted':predictions})
plt.figure(figsize=(8, 6))
plt.scatter(result_df['Actual'], result_df['Predicted'], alpha=1)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Linear Regression Results')
plt.grid(True)


plt.plot([result_df['Actual'].min(), result_df['Actual'].max()],
         [result_df['Actual'].min(), result_df['Actual'].max()],
         linestyle='--', color='red')

plt.show()
```

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(ya_test, predictions))
print('MSE:', metrics.mean_squared_error(ya_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(ya_test, predictions)))
```
