```python
# Import our dependencies
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
# It suggest sklearn.ensemble.HistGradientBoostingClassifier and Regressor, for NaN handlers, or preprocessing data
```

```python
# Read the CSV file into a Pandas DataFrame
try:
    df_raw = pd.read_csv("CFA_dataset_clean.csv")
except Exception as e:
    print("An error occured:", e)
```

```python
df_raw
```

```python
df_raw['incident_datetime'] = pd.to_datetime(df_raw['incident_datetime'])
```

```python
df_raw
```

```python
enc = OneHotEncoder(handle_unknown='ignore')
enc_arr = enc.fit_transform(df_raw[['Ball_Direction of maximum wind gust']])
column_names = enc.get_feature_names_out(input_features=['Ball_Direction of maximum wind gust'])

enc_df = pd.DataFrame(enc_arr.toarray(), columns=column_names)

CFA_whole_merge = df_raw.join(enc_df)
CFA_whole_merge
```

```python
enc = OneHotEncoder(handle_unknown='ignore')
enc_arr = enc.fit_transform(df_raw[['Fern_Direction of maximum wind gust']])
column_names = enc.get_feature_names_out(input_features=['Fern_Direction of maximum wind gust'])

enc_df = pd.DataFrame(enc_arr.toarray(), columns=column_names)

CFA_whole_merge = df_raw.join(enc_df)
CFA_whole_merge
```

```python
enc = OneHotEncoder(handle_unknown='ignore')
enc_arr = enc.fit_transform(df_raw[['Wang_Direction of maximum wind gust']])
column_names = enc.get_feature_names_out(input_features=['Wang_Direction of maximum wind gust'])

enc_df = pd.DataFrame(enc_arr.toarray(), columns=column_names)

CFA_whole_merge2 = df_raw.join(enc_df)
CFA_whole_merge2
```

```python
CFA_whole_merge2.drop(columns=['Ball_Direction of maximum wind gust','Fern_Direction of maximum wind gust', 'Wang_Direction of maximum wind gust'], inplace=True)
```

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt



# Extract numeric features (e.g., year, month, day) for ACF and PACF analysis
CFA_whole_merge2['year'] = CFA_whole_merge2['incident_datetime'].dt.year
CFA_whole_merge2['month'] = CFA_whole_merge2['incident_datetime'].dt.month
CFA_whole_merge2['day'] = CFA_whole_merge2['incident_datetime'].dt.day

# Create ACF and PACF plots 
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))

# ACF plot
sm.graphics.tsa.plot_acf(df_raw['Incident_count'], lags=40, ax=ax1)
ax1.set_title('Autocorrelation Function (ACF)')

# PACF plot
sm.graphics.tsa.plot_pacf(df_raw['Incident_count'], lags=40, ax=ax2)
ax2.set_title('Partial Autocorrelation Function (PACF)')

plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
```

```python
data = CFA_whole_merge2
data['incident_datetime'] = pd.to_datetime(data['incident_datetime'])
data.set_index('incident_datetime', inplace=True)


train_data = data.loc['2018-12-01':'2021-12-31']
test_data = data.loc['2022-01-01':'2022-06-30']

# Determine the ARIMA model order (p, d, q) based on ACF and PACF analysis
p, d, q = (1, 0, 0)  # Replace with the appropriate values

arima_model = ARIMA(train_data, order=(p, d, q))
arima_results = arima_model.fit()

predictions = arima_results.forecast(steps=len(test_data))

mse = ((test_data - predictions) ** 2).mean()
rmse = mse**0.5
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")

# Plot the actual vs. predicted values
plt.figure(figsize=(12, 6))
plt.plot(test_data, label='Actual')
plt.plot(predictions, color='red', label='Predicted')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()

```
